```python
import math as m
import numpy as np
import csv
mo_phong=[]
for i in range(0,180,1):
  arr=[] 
  for j in range(0,180,1):
      for k in range(0,180,1):
        x=50*m.cos(m.radians(i))+40*m.cos(m.radians(i+j))+20*m.cos(m.radians(i+j+k))
        y=50*m.sin(m.radians(i))+40*m.sin(m.radians(i+j+k))+20*m.sin(m.radians(i+j+k))
        si=i+j+k
        arr.append(np.array((i,j,k,x,y)))
  mo_phong.extend(arr)
```

```python
header = ['tt1', 'tt2','tt3', 'Px', 'Py',]
with open('Robot3bac.csv', 'w', encoding='UTF8', newline='') as data_table:
    writer = csv.writer(data_table)
    writer.writerow(header)
    writer.writerows(mo_phong)
```

```python
import pandas as pd
data = pd.read_csv ('Robot3bac.csv')
data.head()
```

```python
x=data.iloc[:,[3,4]]
y=data.iloc[:,[0,1,2]]
```

```python
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
cal_names = list(x.columns)
scaler = StandardScaler()
x_df = scaler.fit_transform(x)
x_df = pd.DataFrame(x_df,columns=cal_names)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_df,y,test_size=0.25,random_state=7)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping

model=Sequential() 
model.add(Dense(20,kernel_initializer='normal', activation='relu',input_shape=(2,)))
model.add(Dense(20,activation='relu')) 
model.add(Dense(20,activation='relu')) 
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
x_train_scale=scale(x_train)
scaler=StandardScaler().fit(x_train)
x_test_scale=scaler.transform(x_test)
```

```python
model.compile(loss='mse',metrics=['mean_absolute_error'])
history = model.fit(x_train_scale,y_train,epochs=10,verbose=1,validation_split=0.2)
```

```python
score=model.evaluate(x_test_scale,y_test,verbose=0)
print('Test loss',score[0])
print('Test accuracy',score[1])
```
